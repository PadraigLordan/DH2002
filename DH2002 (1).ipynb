## Steps and Sleep Hours Over Time

This line graph displays your daily step count (black) and sleep hours (green) over a 30-day period.


Tracking both activity and sleep together helps you spot patternsâ€”such as whether high-activity days lead to more or less sleep. This can reveal how your routines impact your rest and overall well-being.

```python
plt.figure(figsize=(12, 6))
plt.plot(personal_data['date'], personal_data['steps'], label='Steps', marker='o', color='black')
plt.plot(personal_data['date'], personal_data['sleep_hours']*2000, label='Sleep Hours (scaled)', marker='x', color='green')
plt.title('Steps and Sleep Hours Over Time')
plt.xlabel('Date')
plt.ylabel('Steps / Sleep Hours (scaled)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Mood Distribution

This histogram shows how often you rated your mood from 1 to 5 each day.


Understanding the distribution of your mood scores helps you reflect on your overall emotional state during the month. It can also help identify periods of stress or positivity, and can be compared to other habits like sleep or activity

```python
plt.figure(figsize=(8, 4))
plt.hist(personal_data['mood'], bins=5, color='red', edgecolor='black')
plt.title('Mood Distribution')
plt.xlabel('Mood (1-5)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Pair Plot

This pairplot displays scatter plots for every pair of tracked variables along with histograms for each individual variable on the diagonal. Each scatter plot lets you visually inspect if two variables are related.

By examining these relationships, we can spot trends, clusters, or outliers in your personal data. This helps discover which habits might be connectedâ€”such as whether increased activity or sleep correlates with better mood, or if more social media use is linked to higher screen time.

```python
sns.pairplot(personal_data.drop(columns=['date']))
plt.suptitle('Pairplot of Personal Data Variables', y=1.02)
plt.show()
```

## Narrative

I started a data collection trip to learn more about myself during the previous month. I tracked five important components of my everyday life: screen time, mood, social media usage, sleep length, and physical activity (steps).  What started out as a straightforward project turned into an intriguing investigation of my behaviors and how they interact.  By exposing trends I hadn't previously seen, the visualizations served as a mirror for my actions and offered insights that could potentially assist me in making better decisions regarding my daily schedule.

####Collection process
From March 20 to April 18, 2025, I painstakingly documented data points that represented various aspects of my existence for 30 days in a row.  I made a record of how many hours I had slept the night before every morning.  My phone monitored my movements throughout the day.  I recorded my total screen time hours, the number of minutes I spent on social media, and my mood on a scale of 1 to 5 before going to bed.  Even if it was sometimes tiresome, the act of measuring itself turned into a ritual that increased my awareness of various facets of my existence.

####What the data revealed
Examining the relationship between mood and other factors yielded the most startling discovery.  Screen time hours had a significant negative association with my mood (-0.43), as the bar chart illustrates, indicating that days with higher screen time were linked to poorer mood ratings.  On the other hand, minutes spent on social media had a positive correlation (0.24) with mood, suggesting that the effects of social media use may differ from those of total screen time.

My original prediction that physical activity and sleep would be powerful indicators of my emotional state was challenged when I found that sleep hours and steps had rather weak associations with mood (0.02 and 0.01 respectively).  This discrepancy between my expectations and the evidence became a recurrent topic in my investigation.

Deeper understanding of the distributions and correlations between the variables was made possible by the pairplot matrix.  My daily step count ranged widely, ranging from roughly 3,400 to 14,700, with the majority of my days falling between 6,000 and 12,000 steps.  Sleep duration was 5-8 hours, with a rare exception of 9 hours.  On a scale of 1 to 5, my mood ratings were reasonably evenly dispersed, with a small predilection for middle values (2-4).  While social media usage varied greatly between 10 and 175 minutes every day, screen time regularly ranged between 1.5 and 6 hours.

####Screen-Time/Mood Correlation
Recent studies on digital use are consistent with the unfavorable relationship between screen time and mood.  "As much as three hours per day of media usage by school-aged children results in a lessened cortisol surge an hour after waking up, which is detrimental to their development," according to one study, and lower cortisol levels lead to health issues.  Even though I'm an adult, the physiological effects of screen time probably exhibit comparable trends.

"Youth who spent the most time on their digital technology were statistically more likely to exhibit higher levels of internalizing problems two years later," such as anxiety and depression, according to another study.  These results appear to be supported by my data, which shows that days when I used screens for more than five hours were consistently associated with poorer mood ratings (1-2).

The fact that this association wasn't immediately apparent until I displayed the data shocked me.  I frequently excused my excessive screen time on days when I needed to work or unwind, not really relating it to how I was feeling at the time.  My subjective experience had overlooked the objective perspective that the data visualization produced.

####Social Media Paradox
Interestingly, social media minutes showed a favorable link with mood, but overall screen time revealed a negative correlation.  Given that social media offers a distinct form of involvement from other screen activities, this seeming contradiction may help to explain why I still use it in spite of worries about screen time.

 The scatter plots, however, show that this relationship isn't simple.  The wide range of responses raises the possibility that the type or content of social media interaction is more important than the amount of time spent on it.  While extended stretches of passive scrolling had little effect on some days, short but meaningful connections may have improved my happiness on others.

 ####Physical Activity and Sleep: Surprising Non-Correlations
 The most surprising discovery was probably that there was almost no relationship between mood and physical activity, despite a lot of evidence to the contrary.  A single study found that "physical activity may be closely related to psychological resilience and may further influence sleep quality through psychological resilience" .  Another study discovered that "exercise has a negative correlation with rumination, depressive mood, and sleep quality" .

 Why did these well-established trends not appear in my data?  It's possible that the measures are too imprecise, the time period is too brief, or my unique reaction to exercise deviates from the norm.  Furthermore, the link may be more nuanced than a straightforward correlation; for example, physical activity may improve mood only in conjunction with other factors or with a delayed effect.

 ####Living With Data: From Insights to Changes
 Based on preliminary trends, I found myself making minor behavioral tweaks as the days went by and I started examining my developing dataset.  I became more deliberate about my digital intake after realizing the link between screen time and mood, sometimes choosing non-screen activities when I saw excessive usage.  According to study, "Self-monitoring engages the users, encourages them to reflect on their behavior and thus enables them to make informed decisions." This real-time feedback loop is one of the most potent features of self-tracking.

 But there were drawbacks to the tracking procedure itself.  There were days when I felt a subtle strain that was "oppressive and 'punishy'" since I knew I would be recording my actions.  The data's interpretation was confounded by a meta-layer of behavioral impact, as I periodically caught myself taking late-night steps to boost my count or feeling bad about using screens.

 ####Integration With Lived Experience
Comparing the objective statistics with my subjective experience was the most beneficial part of this endeavor.  Despite taking a lot of steps (14,087) and using social media sparingly (10 minutes), but spending a lot of time on screens (4.7 hours), I registered my lowest mood (1) on March 28.  My diary entry from that day highlighted work-related stress from an extended video conference, which was a qualitative detail that the stats alone were unable to convey.

 In a similar vein, I noted peak mood ratings (5) on March 31 and April 6 for two rather different behavioral profiles: the former had a high step count and sleep, while the latter had a moderate step count but no screen time.  These days lined up with significant personal occasionsâ€”a family get-together and a hiking trip, respectivelyâ€”showing how context and meaning impact wellbeing in ways that go beyond quantifiable actions.

 ####Recommendations for Future Self-Tracking
 If I were to continue or repeat this self-tracking experiment, I would make several adjustments:
Track the quality, not just quantity, of behaviors (particularly sleep and screen time)

1.Include measurements of stress levels and social interaction

2.Record contextual notes about significant events or external factors

3.Experiment with interventions based on emerging patterns

4.Extend the timeframe to capture longer-term trends and seasonal effects

This process has given me the "deeper understanding of user behavior," "personalized experiences using behavioral insights," and "improved strategic planning with data-driven" decisions that behavioral databases provide to organizations. However, the user in this process is me, and the insights inform my personal decisions rather than business plans.

####Conclusion: The Value of Personal Data Narratives
This 30-day self-tracking experiment has reaffirmed that statistics only provide a portion of the picture.  To bring about significant change, the correlations and patterns must be combined with subjective experience, even though they offer insightful objective information.  Regardless of particular results, the process itself raised awareness of routines and decisions that normally function subconsciously.

According to William James, "The more details of our daily life we can hand over to the effortless custody of automatism, the higher mental powers of mind will be set free for their own proper work" .  Ironically, by using monitoring to momentarily bring automatic actions into conscious awareness, I can eventually form healthier routines that can again become automatic, but this time they will be purposefully created rather than unintentionally picked up.

A suitable metaphor for the intricate interplay between technology, self-awareness, and wellbeing in our digital age, the screen in your hand as you read this story serves as both a tool for insight and possibly a component of the pattern under investigation.
